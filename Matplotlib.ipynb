## Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 6)) # width, height

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple lable",
       xlabel="x-axis",
       ylabel="y-label")

# 5. Save and show (you save the whole figure)
fig.savefig("Screenshot (2).png")
```

# Making figures with NumPy arrays

We want:

* .plot() = Line plot
* .scatter() = Scatter plot
* .bar() or .barh(list()) = Bar plot or horizontal bar
* .hist() = Histogram
* .subplots() = Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use the same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Jack's butter",
       ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

# Two options for subplots

```python
# Subplots option one
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5));

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option two
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

```python
df = pd.read_csv("heart_disease.csv")
df
```

```python
#
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=df["age"],
                      y=df["chol"],
                      c=df["thalach"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="target")

# Add a mean line
ax0.axhline(y=df["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=df["age"],
                      y=df["thalach"],
                      c=df["thalach"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="age",
        ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="target")

# Add a mean line
ax1.axhline(y=df["thalach"].mean(),
            linestyle="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```
